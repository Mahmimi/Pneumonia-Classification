## Full code

```python
from PIL import Image, ImageEnhance, ImageOps
import numpy as np

# Load the image
image = Image.open("D:\\programing\\Pneumonia\\Images\\Chest_xray\\2D\\train\\NORMAL\\IM-0329-0001.jpeg")
image = image.resize((224, 224))
np_image = np.array(image)

# Shift the image horizontally by a few pixels to simulate depth
shifted_image = np.roll(np_image, shift=5, axis=1)  # Shift right by 5 pixels

# Convert back to an image
shifted_image = Image.fromarray(shifted_image)

red_img = ImageOps.colorize(image, (0, 0, 0), (255, 0, 0))
cyan_img = ImageOps.colorize(shifted_image, (0, 0, 0), (0, 255, 255))

blend = Image.blend(red_img, cyan_img, 0.5)

brightness = ImageEnhance.Brightness(blend)
image_3d = brightness.enhance(1.5)
image_3d
```
